Сравнение групп А/B-эксперимента
Для каждого события проверим с помощью z-критерия гипотезу о равенстве долей пользователей групп 247 и 248, совершивших эти события.

Сформулируем гипотезы:

Нулевая гипотеза  𝐻0:
  Для заданного события разница между долями пользователей двух исследуемых групп, совершивших это событие, равна 0;

Альтернативная гипотеза  𝐻1:
  Для заданного события разница между долями пользователей двух исследуемых групп, совершивших это событие, не равна 0.

В данном случае под "исследуемыми группами" будем понимать группы 247 и 248.

```python
def z_test(group1, group2, event, alpha):
    p1 = p1_ev / p1_us
    p2 = p2_ev / p2_us
    difference = p1 - p2
    p_combined = (p1_ev + p2_ev) / (p1_us + p2_us)
    z_value = difference / math.sqrt(p_combined * (1 - p_combined) * (1 / p1_us + 1 / p2_us))
    distr = st.norm(0, 1)
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    print('Группы {} и {}. Событие -- {}. P-значение --  {:.2f}. Уровень значимости -- {}'.format(group1, group2, event, p_value, alpha ))
    if (p_value < alpha):
        print("Отвергаем нулевую гипотезу -- между долями есть значимая разница")
    else:
        print("Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными")
```

```python
#данные берем из датафрейма count_users_groups
group1 = 247 #номер группы
group2 = 248 #номер группы
event = 'MainScreenAppear' #исследуемое событие
alpha = 0.05 #уровень значимости

p1_ev = 2476 #число пользователей группы 247 совершивших событие 'MainScreenAppear'
p2_ev = 2493 #число пользователей группы 248 совершивших событие 'MainScreenAppear'
p1_us = 2512 #общее число пользователей группы 247
p2_us = 2535 #общее число пользователей группы 248

z_test(247, 248, event, 0.05) # проверка гипотезы

#повторяем аналогичные действия для оставшихся событий
```
